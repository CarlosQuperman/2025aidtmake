### 관계 시각화 그래프(네트워크 시각화 그래프)

- 데이터들이 서로 관계를 맺고 있을 때 이를 효과적으로 시각화하는 방법(그래프 시각화)
- Python에서는 Pyvis 라이브러리를 활용하면 데이터를 네트워크 그래프로 쉽게 시각화할 수 있다.
- 가족관계도, SNS 친구 관계, 교우 관계, 버스 운영 노선, 유전 데이터 등이 네트워크 그래프로 시각화하기에 적절한 데이터이다.
- 공간에서 데이터를 나타날 때 공간을 의미하는 점을 정점(Vertex, Node), 간선(Edge)라고 한다.

```python
#노드, 에지 생성

#모듈 불러오기
from pyvis.network import Network
from IPython.display import HTML

#그래프 생성
g = Network(notebook=True, cdn_resources='in_line')

#노드 추가
g.add_node(0)
g.add_node(1)
g.add_node(2)

#간선 추가
g.add_edge(0, 1, arrows='to')
g.add_edge(0, 2, arrows='to')
g.add_edge(1, 2, arrows='to')


#그래프의 정점과 간선 정보 출력
for n in g.nodes:
  print(n)
for n in g.edges:
  print(n)

#그래프 파일 저장 및 보여주기
g.save_graph("g1.html")
HTML(filename="g1.html")
```

```python
#노드에 레이블 추가
##label에 임의의 값 집어넣기

g = Network(notebook=True, cdn_resources='in_line')
g.add_node(0, label='')
g.add_node(1, label='')
g.add_edge(0, 1, arrows='to')

#그래프의 정점과 간선 정보 출력
for n in g.nodes:
  print(n)

for n in g.edges:
  print(n)

#그래프 내용을 파일로 저장하고, 그 파일을 불러오기
g.save_graph("g2.html")
HTML(filename="g2.html")
```

```python
#노드에 툴팁(마우스로 클릭했을 때 나오는 글자 )추가하기

from pyvis.network import Network

#정점 정보 추가
## label과 title에 임의의 값 집어넣고 결과 확인하기
g = Network(notebook=True, cdn_resources='in_line')
g.add_node(0, label='', title='')
g.add_node(1, label='', title='')
g.add_node(2, label='', title='')

#간선 정보 추가
g.add_edge(0, 1, arrows='to')
g.add_edge(0, 2, arrows='to')

#그래프의 정점과 간선 정보 출력
for n in g.nodes:
  print(n)
for n in g.edges:
  print(n)

g.save_graph("g3.html")
HTML(filename="g3.html")
```

```python
#노드 모양, 크기, 색상, 그림자 변경

from pyvis.network import Network


g = Network(notebook=True, cdn_resources='in_line')

#정점 정보 생성
g.add_node(0, label='A', title='I am A', shape='diamond')
g.add_node(1, label='B', title='I am B', shape='box')
g.add_node(2, label='C', title='new C', shape='triangle', size=50)
g.add_node(3, label='D', shape='triangleDown')
g.add_node(4, label='E', shape='square', color='red')
g.add_node(5, label='F', shape='database', color='green', shadow=True)

g.add_node(6, label='정보', shape='square', color='#4b0082', shadow=True)
g.add_node(7, label='정보과학', shape='database', color='#fffacd', shadow=True)
g.add_node(8, label='프로그래밍', shape='database', color='#add8e6', shadow=True)

#간선 정보 생성
g.add_edge(0, 1, arrows='to')
g.add_edge(0, 2, arrows='to')
g.add_edge(7, 6, arrows='to')
g.add_edge(8, 6, arrows='to')




#그래프의 정점과 간선 정보 출력
for n in g.nodes:
  print(n)
for n in g.edges:
  print(n)

g.save_graph("g4.html")
HTML(filename="g4.html")

```

```python
#리스트를 활용한 네트워크 그래프 생성 및 설정 박스 생성

net = Network(notebook=True, cdn_resources='in_line')

#정점과 간선을 만들기 위한 정보를 담고 있는 리스트 rels
rels = [

    ["Fred", "George"],
    ["Harry", "Rita"],
    ["Fred", "Rita"],
    ["Fred", "Ginny"],
    ["Tom", "Ginny"],
    ["Harry", "Ginny"],
    ["장원재", "조용수"],
    ["조용수", "장원재"],

]

#리스트 rels 순회하면서 정점 및 간선 설정
for rel in rels:
    source, dest = rel
    net.add_node(source)
    net.add_node(dest)
    net.add_edge(source, dest)

#정점 및 간선 정보 출력
for n in net.nodes:
  print(n)

for n in net.edges:
  print(n)

#그래프 설정 UI 추가

net.show_buttons(filter_=['physics'])

net.save_graph("simple_graph.html")
HTML(filename="simple_graph.html")
```
